```python
import numpy as np

from graspy.plot import heatmap, gridplot
from graspy.simulations import weighted_sbm
%matplotlib inline
```

## Simulate graphs using weighted stochastic block models
The 2-block model is defined as below:
\begin{align*}
    P = \begin{bmatrix}
        0.8 & 0.2 \\
        0.2 & 0.8 
    \end{bmatrix}
\end{align*}

We generate two weight SBMs where the weights are distributed from a Poisson(3) and Normal(5, 1).

```python
n_communities = [50, 50]
P = np.array([[0.8, 0.2], [0.2, 0.8]])
Wt = np.random.poisson
Wtargs = dict(lam=3)
A_poisson= weighted_sbm(n_communities, P, Wt=Wt, Wtargs=Wtargs)

Wt = np.random.normal
Wtargs = dict(loc=0, scale=1)
A_normal = weighted_sbm(n_communities, P, Wt=Wt, Wtargs=Wtargs)
```

## Plot the simulated weighted SBMs using plot_heatmap function

```python
title = 'Weighted Stochastic Block Model with Poisson(3)'

fig= heatmap(A_poisson, title=title, transform=None)
fig= heatmap(A_poisson-1, title=title, transform=None, cmap='RdBu_r')
```

```python
title = 'Weighted Stochastic Block Model with Normal(5, 1)'

# fig= heatmap(A_normal, title=title, transform=None, cmap='RdBu_r')
import seaborn as sns
sns.heatmap(A_normal + 10, cmap='Blues', center=0)
```

## Simulate more graphs using weighted stochastic block models
The 2-block model is defined as below:
\begin{align*}
    P = \begin{bmatrix}
        0.25 & 0.05 \\
        0.05 & 0.25
    \end{bmatrix}
\end{align*}

We generate two weighted SBMs where the weights are distributed from a discrete uniform(1, 10) and discrete uniform(2, 5).

```python
n_communities = [50, 50]
P = np.array([[0.25, 0.05], [0.05, 0.25]])
Wt = np.random.randint
Wtargs = dict(low=1, high=10)
A_unif1= weighted_sbm(n_communities, P, Wt=Wt, Wtargs=Wtargs)

Wtargs = dict(low=2, high=5)
A_unif2= weighted_sbm(n_communities, P, Wt=Wt, Wtargs=Wtargs)
```

## Plot the pair of graphs using plot_grid_plot function

```python
X = [A_unif1, A_unif2]
labels = ["Uniform(1, 10)", "Uniform(2, 5)"]

f = gridplot(X=X, labels=labels, title='Two Weighted Stochastic Block Models', height=12, font_scale=1.5, alpha=0.5)
```

### You can also plot one graph if you wanted to

```python
X = A_unif1
labels = ["Uniform(1, 10)"]

f = gridplot(X=[X], labels=labels, 
                   title='One Weighted Stochastic Block Model', height=12, font_scale=1.25)
```
